```python
import cv2
import os

image_frame = cv2.imread('/content/celebrity-face/celebrity face dataset/Ravindra Jadeja/Ravindra Jadeja (10).jpg')
face_detector = cv2.CascadeClassifier('/content/haarcascade_frontalcatface.xml')

faces = face_detector.detectMultiScale(image_frame, 1.3, 5)

for (x,y,w,h) in faces:
    cv2.rectangle(image_frame, (x,y), (x+w,y+h), (0, 0, 0), 0)
    cv2.imwrite("/content/sample_data/test.jpg", image_frame[y:y+h,x:x+w])
```

```python
names = os.listdir("/content/celebrity-face/cropped_cropped_KBS_color_RGB/cropped_cropped_KBS_color_RGB/")
images = []
x = []
y = []
for folder in names:
  files = os.listdir("/content/celebrity-face/cropped_cropped_KBS_color_RGB/cropped_cropped_KBS_color_RGB/"+folder)
  for file in files:
    img = image.load_img("/content/celebrity-face/cropped_cropped_KBS_color_RGB/cropped_cropped_KBS_color_RGB/"+folder+"/"+file,target_size=(100, 100))
    images.append(img)
    img = image.img_to_array(img)
    x.append(img)
    y.append(names.index(folder))

x = np.array(x)
y = np.array(y)
print(x.shape)
print(y.shape)
y = to_categorical(y)
print(y.shape)
x = x/255
from tensorflow.keras.applications import resnet50
from tensorflow.keras.callbacks import EarlyStopping
callback = EarlyStopping(monitor = 'val_loss',min_delta = 0,patience = 3, verbose = 1,restore_best_weights = True)
model2 = resnet50.ResNet50(weights='imagenet')
#new_model.summary()
input_layer = layers.Input(shape=(100, 100,3))
model2 = resnet50.ResNet50(weights='imagenet',include_top=False,input_tensor=input_layer)
#model2.summary()
last_layer = model2.output
flat = layers.Flatten()(last_layer)
output_layer = layers.Dense(3,activation='softmax')(flat)
resnet50_model = models.Model(inputs=input_layer,outputs=output_layer)
#new_model.summary()
for layer in resnet50_model.layers[:-1]:
 layer.trainable=False
#new_model.summary()
from sklearn.model_selection import train_test_split
x_Train,x_Test,y_Train,y_Test = train_test_split(x,y,test_size=0.2,random_state=5)
resnet50_model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
resnet50_model.fit(x_Train,y_Train,batch_size=32,epochs=50,validation_data=(x_Test,y_Test), callbacks=[callback])
```
